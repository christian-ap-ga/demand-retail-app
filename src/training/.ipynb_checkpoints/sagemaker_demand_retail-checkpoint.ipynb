# Demand Retail — Entrenamiento e Inferencia con SageMaker BYOC

Este notebook ejecuta el flujo completo de Bring Your Own Container (BYOC) en Amazon SageMaker:
1. Build del container y push a ECR
2. Subida de datos de entrenamiento a S3
3. Training job con SageMaker Estimator
4. Deploy del endpoint de inferencia en tiempo real
5. Predicciones de prueba
6. Predicciones sobre todo el test set → guardadas en `s3://1c-company-medallion/gold/predictions/`
7. Limpieza del endpoint

### Fuentes de datos (S3)
| Canal | Bucket | Prefix |
|---|---|---|
| training: monthly_sales | `1c-company-medallion` | `silver/monthly_sales/` |
| training: items_features | `1c-company-medallion` | `silver/items_features/` |
| training: categories_features | `1c-company-medallion` | `silver/categories_features/` |
| training: shops_features | `1c-company-medallion` | `silver/shops_features/` |
| output: predicciones | `1c-company-medallion` | `gold/predictions/` |

## 1. Configuración del entorno

````python
import boto3
import os
import numpy as np
import pandas as pd
import sagemaker as sage
from pathlib import Path
from sagemaker import get_execution_role
from sagemaker.serializers import CSVSerializer
from time import gmtime, strftime

# IAM role con permisos de SageMaker + ECR + S3
role = get_execution_role()

# Sesión de SageMaker
sess = sage.Session()

# ── Bucket medallion (fuente de datos y destino de predicciones) ────────────
MEDALLION_BUCKET = "1c-company-medallion"

# Bucket y prefix S3 para artefactos del training job (modelo, logs)
ARTIFACTS_BUCKET = "sagemaker-demand-retail"
prefix           = "byoc-v1"

default_bucket_prefix = sess.default_bucket_prefix
if default_bucket_prefix:
    prefix = f"{default_bucket_prefix}/{prefix}"

# Nombre del container en ECR
IMAGE_NAME = "demand-retail-training"

account = sess.boto_session.client("sts").get_caller_identity()["Account"]
region  = sess.boto_session.region_name
image   = f"{account}.dkr.ecr.{region}.amazonaws.com/{IMAGE_NAME}:latest"

print(f"Role:              {role}")
print(f"Medallion bucket:  {MEDALLION_BUCKET}")
print(f"Artifacts bucket:  {ARTIFACTS_BUCKET}")
print(f"Prefix:            {prefix}")
print(f"Image:             {image}")
````

## 2. Build del container y push a ECR

Ejecuta el script `build_and_push.sh` desde el directorio `training/` del repositorio.
El script construye la imagen Docker y la sube al repositorio ECR `demand-retail-training`.

````python
TRAINING_DIR = os.getcwd()
print(f"Directorio de trabajo: {TRAINING_DIR}")
````

````python
%%sh
chmod +x src/train
chmod +x src/serve
````

````python
%%sh
image="demand-retail-training"

account=$(aws sts get-caller-identity --query Account --output text)
region=$(aws configure get region)
region=${region:-us-east-1}
fullname="${account}.dkr.ecr.${region}.amazonaws.com/${image}:latest"

aws ecr describe-repositories --repository-names "${image}" > /dev/null 2>&1
if [ $? -ne 0 ]; then
    aws ecr create-repository --repository-name "${image}" > /dev/null
fi

aws ecr get-login-password --region "${region}" | \
    docker login --username AWS --password-stdin "${account}.dkr.ecr.${region}.amazonaws.com"

docker build --network sagemaker -t ${image} .
docker tag ${image} ${fullname}
docker push ${fullname}

echo "Image subida: ${fullname}"
````

## 3. Registrar rutas de datos en S3

Los datos ya viven en `s3://1c-company-medallion/silver/`. **No se suben** — solo se registran
los prefijos S3 para pasarlos como canales al Estimator.

Estructura esperada en cada prefijo (un archivo parquet por directorio):
```
silver/monthly_sales/          → monthly_sales.parquet
silver/items_features/         → items_catalog.parquet
silver/categories_features/    → items_categories_catalog.parquet
silver/shops_features/         → shops_catalog.parquet
```

````python
# Prefijos S3 de cada fuente de datos en la capa silver
data_channels = {
    "training": f"s3://{MEDALLION_BUCKET}/silver/monthly_sales/",
    "items":    f"s3://{MEDALLION_BUCKET}/silver/items_features/",
    "cats":     f"s3://{MEDALLION_BUCKET}/silver/categories_features/",
    "shops":    f"s3://{MEDALLION_BUCKET}/silver/shops_features/",
}

for name, path in data_channels.items():
    print(f"  {name:10s} → {path}")
````

## 4. Training job con SageMaker Estimator

SageMaker descarga cada canal S3 en `/opt/ml/input/data/<canal>/` dentro del container.
El script `train` lee los parquets de entrenamiento y catálogos, entrena el modelo y
escribe el artefacto en `/opt/ml/model/` (que SageMaker empaqueta y sube a S3 automáticamente).

Los hiperparámetros `*_file` deben coincidir con el nombre del parquet en cada directorio S3.

````python
s3_output_path = f"s3://{ARTIFACTS_BUCKET}/{prefix}/output"

estimator = sage.estimator.Estimator(
    image_uri=image,
    role=role,
    instance_count=1,
    instance_type="ml.m5.large",
    output_path=s3_output_path,
    sagemaker_session=sess,
    hyperparameters={
        "n_estimators":   "400",
        "learning_rate":  "0.05",
        "max_depth":      "9",
        "random_state":   "42",
        # Nombres de archivo dentro de cada canal S3
        "monthly_file":   "monthly_sales.parquet",
        "items_file":     "items_catalog.parquet",
        "categories_file":"items_categories_catalog.parquet",
        "shops_file":     "shops_catalog.parquet",
    }
)

estimator.fit(data_channels)
print("Training job completado")
````

## 5. Deploy del endpoint de inferencia en tiempo real

SageMaker descarga el artefacto del modelo (`.tar.gz`) de S3, lo descomprime en `/opt/ml/model/`
e invoca el container con el comando `serve` (nginx → gunicorn → Flask).

El endpoint necesita permisos `s3:PutObject` sobre `1c-company-medallion/gold/predictions/`
para persistir las predicciones automáticamente.

````python
predictor = estimator.deploy(
    initial_instance_count=1,
    instance_type="ml.m4.xlarge",
    serializer=CSVSerializer()
)

print(f"Endpoint activo: {predictor.endpoint_name}")
````

## 6. Predicciones en tiempo real

Enviamos un CSV al endpoint con columnas `[ID, shop_id, item_id]`.

El predictor devuelve un CSV con el nuevo schema:
```
id, item_id, shop_id, category_id, region, date, created_at, value, value_upper, value_lower
```
y guarda automáticamente el resultado en:
`s3://1c-company-medallion/gold/predictions/prediction_date=<YYYY-MM-DD>/predictions_<run_id>.parquet`

````python
# Carga el archivo de test — columnas requeridas: [ID, shop_id, item_id]
# El archivo test.csv puede seguir viviendo localmente o descargarse desde S3
BASE_DIR = Path(os.getcwd()).parent.parent
test_df  = pd.read_csv(BASE_DIR / "data/raw/test.csv")
print(f"Registros de test: {len(test_df):,}")
test_df.head()
````

````python
import io

# Prueba rápida con 50 registros
sample     = test_df.head(50)
csv_buffer = io.StringIO()
sample.to_csv(csv_buffer, index=False)

response    = predictor.predict(csv_buffer.getvalue())
predictions = pd.read_csv(io.StringIO(response.decode("utf-8")))

print(f"Predicciones recibidas: {len(predictions)}")
print(f"Columnas: {list(predictions.columns)}")
predictions.head(10)
````

````python
# Predicciones sobre todo el test set en lotes de 500
BATCH_SIZE      = 500
all_predictions = []

for i in range(0, len(test_df), BATCH_SIZE):
    batch = test_df.iloc[i:i + BATCH_SIZE]

    buf = io.StringIO()
    batch.to_csv(buf, index=False)

    response    = predictor.predict(buf.getvalue())
    batch_pred  = pd.read_csv(io.StringIO(response.decode("utf-8")))
    all_predictions.append(batch_pred)

    if i % 5000 == 0:
        print(f"Procesados {i}/{len(test_df)} registros...")

final_predictions = pd.concat(all_predictions, ignore_index=True)

# Copia local opcional
local_out = BASE_DIR / "data/predictions/predictions_realtime.csv"
local_out.parent.mkdir(parents=True, exist_ok=True)
final_predictions.to_csv(local_out, index=False)

print(f"Listo: {len(final_predictions):,} predicciones")
print(f"Columnas: {list(final_predictions.columns)}")
print(f"Guardadas localmente en: {local_out}")
print("Las predicciones también fueron persistidas en S3 por el endpoint:")
print(f"  s3://1c-company-medallion/gold/predictions/prediction_date=<YYYY-MM-DD>/")
````

## 7. Limpieza — eliminar el endpoint

Los endpoints generan costo mientras están activos. Elimínalo cuando termines.

````python
predictor.delete_endpoint()
print(f"Endpoint {predictor.endpoint_name} eliminado")
````
